```python
from __future__ import print_function
__author__= 'ShivamKumarPathak-CUSP2018 '

import pylab as pl
# this package sets up pretty plots
import seaborn
from scipy.optimize import curve_fit, minimize
import numpy as np
    
%pylab inline
```

```python
# Initiate a dictionary for storing all distributions
dict_distributions = {}
```

```python
# List to store sizes of different samples
list_sizes = [] 

# Create different sample sizes
for i in range(100):
    list_sizes.append(int(i*20.11)+10)

# Information about list_sizes
print("Total no. of elements:", len(list_sizes))
print(list_sizes)
```

```python
# Setup numpy seed
np.random.seed(102)
```

```python
# Setup mu: to be used in almost all distribution as an input variable for mean or equivalent attribute
var_mu=100
```

## Normal Distribution

```python
# Initiate a dictionary for 'normal' within the dictionary for distributions 
dict_distributions['normal'] = {} 

# Create normal distribution for different sample sizes
# We will use var_mu which is 100 as input mean for normal distributions 
for n in list_sizes:
    dict_distributions['normal'][n] = np.random.normal(var_mu, size = n)
```

```python
# Initiate a dictionary for 'means' within the dictionary for 'normal' distributions 
dict_distributions['normal']['means'] = {}
```

```python
# Plot means of different sample sizes with the sample size as x-axis
ax_normMuN = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in dict_distributions['normal']:
    if not type(nn) == str:
        dict_distributions['normal']['means'][nn] = dict_distributions['normal'][nn].mean()
        #and plot it
        ax_normMuN.plot(nn, dict_distributions['normal']['means'][nn], 'o')
        ax_normMuN.set_xlabel('sample size', fontsize=18)
        ax_normMuN.set_ylabel('sample mean', fontsize=18)
        ax_normMuN.set_title('Normal', fontsize=18)
        ax_normMuN.plot([min(list_sizes), max(list_sizes)], [var_mu, var_mu], 'k')
```

## Poisson Distribution

```python
# Initiate a dictionary for 'poisson' within the dictionary for distributions 
dict_distributions['poisson'] = {} 

# Create poisson distribution gor different sample sizes
# In poisson distribution mean=lambda: we will use var_mu as lambda 
for n in list_sizes:
    dict_distributions['poisson'][n] = np.random.poisson(var_mu, size = n)
```

```python
# Initiate a dictionary for 'means' within the dictionary for 'poisson' distributions 
dict_distributions['poisson']['means'] = {}
```

```python
# Plot means of different sample sizes with the sample size as x-axis
ax_poissonMuN = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in dict_distributions['poisson']:
    if not type(nn) == str:
        dict_distributions['poisson']['means'][nn] = dict_distributions['poisson'][nn].mean()
        #and plot it
        ax_poissonMuN.plot(nn, dict_distributions['poisson']['means'][nn], 'o')
        ax_poissonMuN.set_xlabel('sample size', fontsize=18)
        ax_poissonMuN.set_ylabel('sample mean', fontsize=18)
        ax_poissonMuN.set_title('Poisson', fontsize=18)
        ax_poissonMuN.plot([min(list_sizes), max(list_sizes)], [var_mu, var_mu], 'k')
```

## Binomial Distribution

```python
# Initiate a dictionary for 'binomial' within the dictionary for distributions 
dict_distributions['binomial'] = {} 

# Create binomial distribution for different sample sizes
# We take p as 0.5
# Mean for binomial = n * p = 100; n=200
var_p = 0.5
var_n = 200

for n in list_sizes:
    dict_distributions['binomial'][n] = np.random.binomial(var_n, var_p, size = n)
```

```python
# Initiate a dictionary for 'means' within the dictionary for 'binomial' distributions 
dict_distributions['binomial']['means'] = {}
```

```python
# Plot means of different sample sizes with the sample size as x-axis
ax_binomialMuN = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in dict_distributions['binomial']:
    if not type(nn) == str:
        dict_distributions['binomial']['means'][nn] = dict_distributions['binomial'][nn].mean()
        #and plot it
        ax_binomialMuN.plot(nn, dict_distributions['binomial']['means'][nn], 'o')
        ax_binomialMuN.set_xlabel('sample size', fontsize=18)
        ax_binomialMuN.set_ylabel('sample mean', fontsize=18)
        ax_binomialMuN.set_title('Binomial', fontsize=18)
        ax_binomialMuN.plot([min(list_sizes), max(list_sizes)], [var_mu, var_mu], 'k')
```

## Chi-Square Distribution

```python
# Initiate a dictionary for 'chisq' within the dictionary for distributions 
dict_distributions['chisq'] = {} 

# Create chisq distribution gor different sample sizes
# For chi-square df = mean; we input var_mu as df
for n in list_sizes:
    dict_distributions['chisq'][n] = np.random.chisquare(var_mu, size = n)
```

```python
# Initiate a dictionary for 'means' within the dictionary for 'chisq' distributions 
dict_distributions['chisq']['means'] = {}
```

```python
# Plot means of different sample sizes with the sample size as x-axis
ax_chisqMuN = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in dict_distributions['chisq']:
    if not type(nn) == str:
        dict_distributions['chisq']['means'][nn] = dict_distributions['chisq'][nn].mean()
        #and plot it
        ax_chisqMuN.plot(nn, dict_distributions['chisq']['means'][nn], 'o')
        ax_chisqMuN.set_xlabel('sample size', fontsize=18)
        ax_chisqMuN.set_ylabel('sample mean', fontsize=18)
        ax_chisqMuN.set_title('Chi squared', fontsize=18)
        ax_chisqMuN.plot([min(list_sizes), max(list_sizes)], [var_mu, var_mu], 'k')
```

## Lognormal Distribution

```python
# Initiate a dictionary for 'lognormal' within the dictionary for distributions 
dict_distributions['lognormal'] = {} 

# Create chisq distribution gor different sample sizes
# For lognormal distribution mean= exp(mu + ((std)**2)/2)
# std=1; mean=100; mu= ln(100)-1/2 = 4.10157
var_logmu = 4.10157
var_std = 1
for n in list_sizes:
    dict_distributions['lognormal'][n] = np.random.lognormal(var_logmu,var_std,n)
```

```python
# Initiate a dictionary for 'means' within the dictionary for 'lognormal' distributions 
dict_distributions['lognormal']['means'] = {}
```

```python
# Plot means of different sample sizes with the sample size as x-axis
ax_lognormalMuN = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in dict_distributions['lognormal']:
    if not type(nn) == str:
        dict_distributions['lognormal']['means'][nn] = dict_distributions['lognormal'][nn].mean()
        #and plot it
        ax_lognormalMuN.plot(nn, dict_distributions['lognormal']['means'][nn], 'o')
        ax_lognormalMuN.set_xlabel('sample size', fontsize=18)
        ax_lognormalMuN.set_ylabel('sample mean', fontsize=18)
        ax_lognormalMuN.set_title('Lognormal', fontsize=18)
        ax_lognormalMuN.plot([min(list_sizes), max(list_sizes)], [var_mu, var_mu], 'k')
```
