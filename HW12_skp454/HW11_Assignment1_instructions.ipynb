Note on preprocessing: the preparation of these data should include the "standardization" of the time seris. 
*Always when you apply a moachine learning model you should standardize the data or whiten it.*
These operations may be complicated preprocessing operation, they can for example invoke PCA, and they are aimed at removing covariance in the feature space.

The simplest version of whitening *standardizes each feature*" from every column (feature) you subtract the mean and you divide it by the standard deviation.

But the time series preprocessing is different if you want to use each time stamp as feature: because you want to maintain the coherence of the data, i.e. you **do not want** to treat each feature as an independent feature! the pattenrs in the time series are what you are clustering on (time series goes up and down, creating a pattern) and those patterns would be lost by whitening.

*So when you cluster or otherwise use a time series in a model, using each timestamp as a feature, you want to standardize the observations, not the features.* 

**From each row of your table (observation, in this case each zipcode) subtract the mean and you divide the row by the standard deviation.**

```python
# Check if PUIDATA is present or not and if not set it up
if os.getenv("PUIDATA")==None:
    os.environ["PUIDATA"] = "%s/PUIdata"%os.getenv("HOME")
    print("Earlier missing now, updated:  PUIdata:", os.getenv("PUIDATA"))
else:
    print("Already present: PUIdata", os.getenv("PUIDATA"))
```

```python
# Download and load NYC Shapefile
url = "https://data.cityofnewyork.us/download/i8iw-xf4u/application%2Fzip"
urllib.request.urlretrieve(url, 'nycshp.zip')
os.system("unzip " + 'nycshp.zip -d' + 'nycshp/' )
os.system('mv ' + 'nycshp ' + os.getenv("PUIDATA") )
gdf_nyc = gpd.read_file(os.getenv("PUIDATA")+'/nycshp/ZIP_CODE_040114.shp')
print("ZIPCODES:",
      len(np.unique(gdf_nyc.ZIPCODE)), "POLYGONS:",
      len(gdf_nyc.ZIPCODE),
      "the difference is due to island with same zip as parts of land,",
      "but disjoint shapes")
nyczips = gdf_nyc.ZIPCODE.values.astype(int)
print("number of NYC zip codes:", len(nyczips))
gdf_nyc.head()
```

```python
# Download the data for business establishments
lis_zipnames = ['94', '95', '96', '97', '98', '99', '00', '01', '02', '03', '04', 
                '05', '06', '07', '08', '09', '10', '11', '12', '13', '14']
for i in range(len(lis_zipnames)):
    url = 'https://github.com/fedhere/PUI2018_fb55/raw/master/HW12_fb55/zbp'+str(lis_zipnames[i])+'totals.zip'
    urllib.request.urlretrieve(url, 'zbp'+str(lis_zipnames[i])+'totals.zip')
    os.system("unzip " + 'zbp'+str(lis_zipnames[i])+'totals.zip -d' + 'zbp'+str(lis_zipnames[i])+'totals/' )
    os.system('mv ' + 'zbp'+str(lis_zipnames[i])+'totals ' + os.getenv("PUIDATA") )
```

```python
# import the data for business establishments
df_main = pd.DataFrame()
for i in range(len(lis_zipnames)):
    fname = os.getenv("PUIDATA")+ '/zbp'+str(lis_zipnames[i])+'totals/' +'zbp'+str(lis_zipnames[i])+'totals.txt'
    df = pd.read_csv(fname)
    if 'ZIP' in df.columns:
        df.rename(columns={'ZIP' : 'zip' , 'EST' : 'est'} , inplace=True)
    df = df[['zip','est']]
    df.drop_duplicates(['zip'], inplace =True)
    df.index = df['zip']
    df.drop('zip', axis=1, inplace=True)
    df.columns = ['est'+lis_zipnames[i]]
    df_main = pd.concat([df_main,df], axis=1)

df_main.head()    
```

```python
# Keep only zipcodes relevant for NYC
df_main = df_main.loc[gdf_nyc['ZIPCODE'].astype(int),:]
print(df_main.shape)
df_main.head()
```

```python
# Drop missing value
df_main.dropna(inplace =True)

# Standardize each row
scaler =  StandardScaler()
df_mainCluster = pd.DataFrame(data = scaler.fit_transform(df_main.values.T).T)

df_mainCluster.describe()
```

### K- Mean Clustering to Cluster Zipcodes based on Time Series of Business Establishments

```python
# k means determine k
X = df_mainCluster
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

#### Fig: Elbow Plot for Kmean clustering
1. In the above plot we can see the decrease in distortion after 2 clusters is relatively low.
2. Another elbow point is 5.
3. We take 5 as optimum number of clusters here as 2 won't be enough to accomodate all patterns.

```python
#cluster with kmeans
km = KMeans(n_clusters=5, random_state=444).fit(X)
```

```python
#each observation in the km object has a label: 
#that is the cluster assigned to that observation
km.labels_.shape
```

```python
# Add labels to main dataframe
columns = df_mainCluster.columns
df_mainCluster['label'] = km.labels_
```

```python
df_labels = pd.DataFrame(data={'zip':df_main.index, 'labels': km.labels_})
df_labels.drop_duplicates('zip', inplace = True)
gdf_nyc['ZIPCODE'] = gdf_nyc['ZIPCODE'].astype(int)
gdf_km = gdf_nyc.merge(df_labels, left_on='ZIPCODE', right_on='zip', how='left')
```

```python
fig = plt.subplots(figsize = (20,10))
for i in df_mainCluster['label'].unique():
    df_temp = df_mainCluster.loc[df_mainCluster['label'] == i, columns]
    plt.subplot(2,3,i+1)
    for j in df_temp.index:
        plt.plot(range(21),df_temp.loc[j,:],'-', color='grey')
    plt.plot(range(21),df_temp.mean(), 'b-',marker='o')
    plt.title('Cluster'+str(i), fontsize=20)
    plt.xticks([])
    plt.yticks([])
plt.show()

fig,ax = plt.subplots(figsize = (15,10))
gdf_km.plot(column = 'labels',ax = ax, cmap = 'Blues_r', legend = True )
plt.xlabel('Longitude (in ft)')
plt.ylabel('Latitude (in ft)')
plt.title('Clusters of Business establishment Time Series')
```

#### Figure: Result of K-Mean Clustering on Time Series of Business establishment in different zip codes.

1. The top 5 plots shows the time series pattern in different clusters formed using k-mean clustering. The blue line shows the mean behavior.

2. The bottom map shows the plot of these NYC clusters on the map of NYC.

3. As we can see above, there is a huge imbalance in the number of zip codes in each clusters. Most of the zip codes are either of type 0 or type 3.

### Gaussian Mixture Model to Cluster Zipcodes based on Time Series of Business Establishments
